```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
salary_data = pd.read_csv("Salary_Data.csv")
salary_data.head()
```

```python
plt.title("Experience Vs Salary")
plt.ylabel('Salary')
plt.xlabel('Experience')
sns.lineplot(x=salary_data['YearsExperience'], y=salary_data['Salary'], c='g')
plt.show()
```
